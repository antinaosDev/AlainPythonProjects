```python
import matplotlib.pyplot as plt
import numpy as np
import read_csv
import pandas as pd 
import Risk_per_region
import Risk_per_year
import Risk_per_town
```

```python
data = read_csv.read_csv('agro_emergency.csv')

data_region = Risk_per_region.risk_region(data)

data_year = Risk_per_year.risk_year(data)

data_comuna = Risk_per_town.risk_comuna(data)

data_comuna
```

```python
labels = data_region['Region']
values = data_region['count_sit_r']

labels2 = data_year['Año']
values2 = data_year['count_sit_year']

labels3 = data_comuna['Comuna']
values3 = data_comuna['count_sit_comuna']


```

```python
plt.subplot(2,1,1)
plt.bar(labels,values)
plt.xticks(rotation =45)

plt.xticks(rotation =45)

plt.subplot(2,1,2)
plt.bar(labels2,values2)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(3,1,figsize =(20,15))

axes[0].bar(labels,values,color=['blue'])
axes[0].set_title('Riesgo historico por Región')
axes[0].set_xlabel('Region')
axes[0].set_ylabel('count_sit_r')

axes[1].barh(labels2,values2,color=['red'])
axes[1].set_title('Riesgo por año')
axes[1].set_xlabel('Año')
axes[1].set_ylabel('count_sit_year')

axes[2].barh(labels3,values3,color=['green'])
axes[2].set_title('Riesgo por año')
axes[2].set_xlabel('Comuna')
axes[2].set_ylabel('count_sit_comuna')

plt.tight_layout()
```
